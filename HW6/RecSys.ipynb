```python
from surprise import Dataset
from surprise import Reader
import os
from surprise import SVD, NMF
from surprise import evaluate, print_perf
#load data from a file 
file_path = os.path.expanduser('restaurant_ratings.txt')
reader = Reader(line_format='user item rating timestamp', sep='\t')
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
results = {}
```

```python
# SVD
algo = SVD()
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results["svd"] = perf
print_perf(perf)
```

```python
algo = SVD(biased=False) 
#PMF
perf = evaluate(algo, data,measures=['RMSE', 'MAE'])
results['pmf'] = perf
print_perf(perf)
```

```python
# NMF
algo = NMF()
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results['nmf'] = perf
print_perf(perf)
```

```python
# userbased CF
from surprise import KNNBasic

algo = KNNBasic(sim_options = {'user_based': True })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results['ucf'] = perf
print_perf(perf)
```

```python
# item based CF
algo = KNNBasic(sim_options = {'user_based': False })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results['icf'] = perf
print_perf(perf)
```

```python
from numpy import mean
names = [key for key in results.keys()]
fold1_rmse = [results[key]['rmse'][0] for key in names]
fold2_rmse = [results[key]['rmse'][1] for key in names]
fold3_rmse = [results[key]['rmse'][2] for key in names]
mean_rmse = [mean(results[key]['rmse']) for key in names]
fold1_mae = [results[key]['mae'][0] for key in names]
fold2_mae = [results[key]['mae'][1] for key in names]
fold3_mae = [results[key]['mae'][2] for key in names]
mean_mae = [mean(results[key]['mae']) for key in names]
```

```python
import matplotlib.pyplot as plt
def plot_bar( y, y_label, title,x=names, x_label="Algorithms"):
    plt.bar(x, y)
    if y_label == "MAE":
        plt.ylim([0.7,0.8])
    else:
        plt.ylim([0.9,1])
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.savefig(title)
    plt.show()
    
plot_bar(y=fold1_mae, y_label="MAE", title="Fold1 MAE")
plot_bar(y=fold2_mae, y_label="MAE", title="Fold2 MAE")
plot_bar(y=fold3_mae, y_label="MAE", title="Fold3 MAE")
plot_bar(y=mean_mae, y_label="MAE", title="Mean MAE")

plot_bar(y=fold1_rmse, y_label="RMSE", title="Fold1 RMSE")
plot_bar(y=fold2_rmse, y_label="RMSE", title="Fold2 RMSE")
plot_bar(y=fold3_rmse, y_label="RMSE", title="Fold3 RMSE")
plot_bar(y=mean_rmse, y_label="RMSE", title="Mean RMSE")



```

```python
# cosine, MSD(Mean Squared Difference), and Pearson similarities 
results_cf = {'ucf':{}, 'icf':{}}
algo = KNNBasic(sim_options = {'name':'MSD','user_based': True })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results_cf['ucf']['msd'] = perf
print_perf(perf)

algo = KNNBasic(sim_options = {'name':'MSD','user_based': False })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results_cf['icf']['msd'] = perf
print_perf(perf)

algo = KNNBasic(sim_options = {'name':'cosine','user_based': True })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results_cf['ucf']['cosine'] = perf
print_perf(perf)

algo = KNNBasic(sim_options = {'name':'cosine','user_based': False })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results_cf['icf']['cosine'] = perf
print_perf(perf)

algo = KNNBasic(sim_options = {'name':'pearson','user_based': True })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results_cf['ucf']['pearson'] = perf
print_perf(perf)

algo = KNNBasic(sim_options = {'name':'pearson','user_based': False })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
results_cf['icf']['pearson'] = perf
print_perf(perf)
```

```python
names = ["msd", "cosine", "pearson"]
vals = [mean(results_cf['icf'][key]['MAE']) for key in names]
plt.title("Similarity Metrics Comparison for ICF (MAE)")
plt.xlabel("Similarity Metric")
plt.ylabel("MAE")
plt.ylim([0.7,0.9])
plt.bar(names, vals)
plt.savefig("Similarity Metrics Comparison for ICF (MAE)")
plt.show()

vals = [mean(results_cf['ucf'][key]['MAE']) for key in names]
plt.title("Similarity Metrics Comparison for UCF (MAE)")
plt.xlabel("Similarity Metric")
plt.ylabel("MAE")
plt.ylim([0.7,0.9])
plt.bar(names, vals)
plt.savefig("Similarity Metrics Comparison for UCF (MAE)")
plt.show()

vals = [mean(results_cf['icf'][key]['RMSE']) for key in names]
print(vals)
plt.title("Similarity Metrics Comparison for ICF (RMSE)")
plt.xlabel("Similarity Metric")
plt.ylabel("RMSE")
plt.ylim([0.9,1.1])
plt.bar(names, vals)
plt.savefig("Similarity Metrics Comparison for ICF (RMSE)")
plt.show()

vals = [mean(results_cf['ucf'][key]['RMSE']) for key in names]
print(vals)
plt.title("Similarity Metrics Comparison for UCF (RMSE)")
plt.xlabel("Similarity Metric")
plt.ylabel("RMSE")
plt.ylim([0.9,1.1])
plt.bar(names, vals)
plt.savefig("Similarity Metrics Comparison for UCF (RMSE)")
plt.show()
```

```python
k_icf = []
k_ucf = []
ks = [num for num in range(20,400,20)]
for num in ks:
    algo = KNNBasic(k = num, sim_options = {'name':'pearson','user_based': False })
    perf = evaluate(algo, data, measures=['RMSE'])
    k_icf.append(mean(perf['RMSE']))
    algo = KNNBasic(k = num, sim_options = {'name':'pearson','user_based': True })
    perf = evaluate(algo, data, measures=['RMSE'])
    k_ucf.append(mean(perf['RMSE']))
```

```python
plt.xlabel("K")
plt.ylabel("RMSE")
plt.title("User Based Collaborative Filtering")
plt.scatter(ks, k_ucf)
plt.show()

plt.xlabel("K")
plt.ylabel("RMSE")
plt.title("Item Based Collaborative Filtering")
plt.scatter(ks, k_icf)
plt.show()

```
